```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1. Load dataset
df = pd.read_csv("C:\\Users\\saket\\Downloads\\titanic.csv")

# Select required features (simple)
inputs = df[['Age', 'Fare']]
target = df['Survived']

# Handle missing values
inputs['Age'].fillna(inputs['Age'].mean(), inplace=True)

# 2. Split dataset
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)

# 3. Create and train model
model = KNeighborsClassifier(n_neighbors=35)
model.fit(X_train, y_train)

# 4. Predict
pred = model.predict(X_test)

# 5. Accuracy
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Load dataset
df = pd.read_csv("C:\\Users\\saket\\Downloads\\titanic.csv")

# Select simple numeric features
inputs = df[['Age', 'Fare']]
target = df['Survived']

# Handle missing values
inputs['Age'].fillna(inputs['Age'].mean(), inplace=True)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)

# Create model
model = GaussianNB()

# Train
model.fit(X_train, y_train)

# Predict
pred = model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Load your CSV file
df = pd.read_csv("C:\\Users\\saket\\Downloads\\kmeanschatgpt.csv")

# We cluster only on VAR1 and VAR2
X = df[['VAR1', 'VAR2']]

# Step 1: Run K-Means with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# Step 2: Predict cluster for the new point
new_point = np.array([[0.906, 0.606]])
cluster_of_new = kmeans.predict(new_point)[0]

print("Predicted cluster for new point:", cluster_of_new)

# Step 3: Majority CLASS in that cluster
cluster_data = df[df['cluster'] == cluster_of_new]
predicted_class = cluster_data['CLASS'].mode()[0]

print("Predicted CLASS:", predicted_class)
```

```python
data = [
    ["medium","skiing","design","single","twenties","no","highRisk"],
    ["high","golf","trading","married","forties","yes","lowRisk"],
    ["low","speedway","transport","married","thirties","yes","medRisk"],
    ["medium","football","banking","single","thirties","yes","lowRisk"],
    ["high","flying","media","married","fifties","yes","highRisk"],
    ["low","football","security","single","twenties","no","medRisk"],
    ["medium","golf","media","single","thirties","yes","medRisk"],
    ["medium","golf","transport","married","forties","yes","lowRisk"],
    ["high","skiing","banking","single","thirties","yes","highRisk"],
    ["low","golf","unemployed","married","forties","yes","highRisk"]
]

# Unconditional probability of 'golf'
total = len(data)
golf_count = sum(1 for row in data if row[1] == "golf")
P_golf = golf_count / total

# Conditional probability P(single | medRisk)
medRisk_rows = [row for row in data if row[6] == "medRisk"]
single_count = sum(1 for row in medRisk_rows if row[3] == "single")
P_single_given_medRisk = single_count / len(medRisk_rows)

print("P(golf) =", P_golf)
print("P(single | medRisk) =", P_single_given_medRisk)
```
